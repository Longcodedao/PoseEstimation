```python
from datasets import MPII
import torch
from torch.utils.data import DataLoader
import torch.nn as nn
import numpy as np
```

```python
inputRes = 256
outputRes = 32

train_dataset = MPII(split = 'train', inputRes = inputRes, outputRes = outputRes)

inp, pts, out = train_dataset[0]
```

```python
valid_dataset = MPII(split = 'valid', inputRes = inputRes, outputRes = outputRes)

inp, pts, out = valid_dataset[0]
```

```python
valid_dataset.plot_image(0)
```

```python
batch_size = 32
num_workers = 2

train_loader = DataLoader(train_dataset, batch_size = batch_size,
                          shuffle = True, num_workers = num_workers)

valid_loader = DataLoader(valid_dataset, batch_size = batch_size,
                          shuffle = True, num_workers = num_workers)
```

```python
batch = next(iter(train_loader))

# len(batch)
images, joints, label = batch
print(f"Shape of Batch images: {images.shape}")
print(f"Joints is: {joints.shape}")
print(f"Shape of Label is: {label.shape}")
```
